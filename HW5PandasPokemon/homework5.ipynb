##**2.4 Max Defense**

```python
def max_defense(collection):
  defense = []
  rank = []
  for i in range(len(df.index)):
    for j in collection :
      if df.loc[:,'name'][i] == j :
        defense.append((df.loc[:,'name'][i],df.loc[:,'defense'][i]))
        rank.append(df.loc[:,'defense'][i])
  print(defense[rank.index(max(rank))])

  return
```

##**2.5 Filter Range**

```python
def filter_range(collection, columns, min, max):
  pokemon = []
  minidx = df[columns].loc[df.loc[:,columns]>=min].apply(int).index
  maxidx = df[columns].loc[df.loc[:,columns]<=max].apply(int).index
  interval = list(minidx & maxidx)
  for i in interval:
    for j in collection:
      if df.loc[:,'name'][i] == j :
        pokemon.append((df.loc[:,'name'].loc[int(i)],df.loc[:,columns].loc[int(i)]))
  print(pokemon)

  return
```

```python
filter_range(collection, 'attack', 20, 40)
```

##**2.6 Mean attack for Type**

```python
def mean_attack_for_type(collection, column1, column2):
  col1 = []
  col2 = []
  for i in range(len(df.index)):
    for j in collection :
      if df.loc[:,'name'][i] == j :
        col1.append(df.loc[:,column1][i])
        col2.append(df.loc[:,column2][i])
  print((sum(col1) + sum(col2))/2/len(col1))

  return
```
